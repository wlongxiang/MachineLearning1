In _multiclass_ logistic regression, the conditional probability of class label $j$ given the image $\bx$ for some datapoint is given by:

$ \log p(t = j \;|\; \bx, \bb, \bW) = \log q_j - \log Z$

where $\log q_j = \bw_j^T \bx + b_j$ (the log of the unnormalized probability of the class $j$), and $Z = \sum_k q_k$ is the normalizing factor. $\bw_j$ is the $j$-th column of $\bW$ (a matrix of size $784 \times 10$) corresponding to the class label, $b_j$ is the $j$-th element of $\bb$.

Given an input image, the multiclass logistic regression model first computes the intermediate vector $\log \bq$ (of size $10 \times 1$), using $\log q_j = \bw_j^T \bx + b_j$, containing the unnormalized log-probabilities per class. 

The unnormalized probabilities are then normalized by $Z$ such that $\sum_j p_j = \sum_j \exp(\log p_j) = 1$. This is done by $\log p_j = \log q_j - \log Z$ where $Z = \sum_i \exp(\log q_i)$. This is known as the _softmax_ transformation, and is also used as a last layer of many classifcation neural network models, to ensure that the output of the network is a normalized distribution, regardless of the values of second-to-last layer ($\log \bq$)

**Warning**: when computing $\log Z$, you are likely to encounter numerical problems. Save yourself countless hours of debugging and learn the [log-sum-exp trick](https://www.xarg.org/2016/06/the-log-sum-exp-trick-in-machine-learning/ "Title").

The network's output $\log \bp$ of size $10 \times 1$ then contains the conditional log-probabilities $\log p(t = j \;|\; \bx, \bb, \bW)$ for each digit class $j$. In summary, the computations are done in this order:

$\bx \rightarrow \log \bq \rightarrow Z \rightarrow \log \bp$

Given some dataset with $N$ independent, identically distributed datapoints, the log-likelihood is given by:

$ \mathcal{L}(\bb, \bW) = \sum_{n=1}^N \mathcal{L}^{(n)}$

where we use $\mathcal{L}^{(n)}$ to denote the partial log-likelihood evaluated over a single datapoint. It is important to see that the log-probability of the class label $t^{(n)}$ given the image, is given by the $t^{(n)}$-th element of the network's output $\log \bp$, denoted by $\log p_{t^{(n)}}$:

$\mathcal{L}^{(n)} = \log p(t = t^{(n)} \;|\; \bx = \bx^{(n)}, \bb, \bW) = \log p_{t^{(n)}} = \log q_{t^{(n)}} - \log Z^{(n)}$

where $\bx^{(n)}$ and $t^{(n)}$ are the input (image) and class label (integer) of the $n$-th datapoint, and $Z^{(n)}$ is the normalizing constant for the distribution over $t^{(n)}$.

Write a function `logreg_gradient(x, t, w, b)` that returns the gradients $\nabla_{\bw_j} \mathcal{L}^{(n)}$ (for each $j$) and $\nabla_{\bb} \mathcal{L}^{(n)}$, i.e. the first partial derivatives of the log-likelihood w.r.t. the parameters $\bW$ and $\bb$, evaluated at a single datapoint (`x`, `t`).
The computation will contain roughly the following intermediate variables:

$
\log \bq \rightarrow Z \rightarrow \log \bp\,,\, \mathbf{\delta}^q
$

followed by computation of the gradient vectors $\nabla_{\bw_j} \mathcal{L}^{(n)}$ (contained in a $784 \times 10$ matrix) and $\nabla_{\bb} \mathcal{L}^{(n)}$ (a $10 \times 1$ vector).

For maximum points, ensure the function is numerically stable.


```python
# 1.1.2 Compute gradient of log p(t|x;w,b) wrt w and b
def logreg_gradient(x, t, w, b):
    
    # YOUR CODE HERE
    logq = x@w + b
    a = np.max(logq)
    logZ = a + np.log(np.sum(np.exp(logq-a)))
    logp = logq - logZ
    j = np.arange(0, 10)
    #print(j)
    #print(t)
    ind = np.where(j == t[0], 1, 0)
    #print(ind)
    dq = ind - np.exp(logq)/np.exp(logZ)
    dL_dw = x.T @ dq
    dL_db = dq
    # here the statement contains logp[:,t] where logp is meant tas a matrix of shape 1x10
    
    return logp[:,t].squeeze(), dL_dw, dL_db.squeeze()
```


### 1.1.3 Stochastic gradient descent (15 points)

Write a function `sgd_iter(x_train, t_train, w, b)` that performs one iteration of stochastic gradient descent (SGD), and returns the new weights. It should go through the trainingset once in randomized order, call `logreg_gradient(x, t, w, b)` for each datapoint to get the gradients, and update the parameters **using a small learning rate of `1e-4`**. Note that in this case we're maximizing the likelihood function, so we should actually performing gradient ___ascent___... For more information about SGD, see Bishop 5.2.4 or an online source (i.e. https://en.wikipedia.org/wiki/Stochastic_gradient_descent)

```python
def sgd_iter(x_train, t_train, W, b):
    # YOUR CODE HERE
    # raise NotImplementedError()
    logp_train = 0
    rate = 1e-4
    index = np.arange(0, x_train.shape[0])
    np.random.shuffle(index)
    for i in range(len(index)):
        x = x_train[index[i]:index[i]+1:,:]
        t = t_train[index[i]:index[i]+1]
        logp,dw,db = logreg_gradient(x,t,W,b)
        logp_train += logp
        W += rate * dw
        b += rate * db
    return logp_train, W, b
```

```python
def forward_pass_single(x, t, w, b):
    number_of_classes = len(b)
    b = np.expand_dims(b, axis=1)
    logq = w.T @ x.T + b
    # log_q is of shape 10 x 1
    q = np.exp(logq)
    Z = np.sum(q)
    # logZ is a log of sum: such as logZ = log(exp(log(q1)+log(q1), ...+ log(qk)))
    # according to this: https://www.xarg.org/2016/06/the-log-sum-exp-trick-in-machine-learning/
    logZ_offset = np.max(logq)
    logZ = logZ_offset + np.log(np.sum(np.exp(logq-logZ_offset)))
    logp = logq - logZ
    # here the statement contains logp[:,t] where logp is meant tas a matrix of shape 1x10
    logp = logp.T
    return logp[:,t].squeeze()
```

# Part 2. Multilayer perceptron


You discover that the predictions by the logistic regression classifier are not good enough for your application: the model is too simple. You want to increase the accuracy of your predictions by using a better model. For this purpose, you're going to use a multilayer perceptron (MLP), a simple kind of neural network. The perceptron will have a single hidden layer $\bh$ with $L$ elements. The parameters of the model are $\bV$ (connections between input $\bx$ and hidden layer $\bh$), $\ba$ (the biases/intercepts of $\bh$), $\bW$ (connections between $\bh$ and $\log q$) and $\bb$ (the biases/intercepts of $\log q$).

The conditional probability of the class label $j$ is given by:

$\log p(t = j \;|\; \bx, \bb, \bW) = \log q_j - \log Z$

where $q_j$ are again the unnormalized probabilities per class, and $Z = \sum_j q_j$ is again the probability normalizing factor. Each $q_j$ is computed using:

$\log q_j = \bw_j^T \bh + b_j$

where $\bh$ is a $L \times 1$ vector with the hidden layer activations (of a hidden layer with size $L$), and $\bw_j$ is the $j$-th column of $\bW$ (a $L \times 10$ matrix). Each element of the hidden layer is computed from the input vector $\bx$ using:

$h_j = \sigma(\bv_j^T \bx + a_j)$

where $\bv_j$ is the $j$-th column of $\bV$ (a $784 \times L$ matrix), $a_j$ is the $j$-th element of $\ba$, and $\sigma(.)$ is the so-called sigmoid activation function, defined by:

$\sigma(x) = \frac{1}{1 + \exp(-x)}$

Note that this model is almost equal to the multiclass logistic regression model, but with an extra 'hidden layer' $\bh$. The activations of this hidden layer can be viewed as features computed from the input, where the feature transformation ($\bV$ and $\ba$) is learned.

## 2.1 Derive gradient equations (20 points)

State (shortly) why $\nabla_{\bb} \mathcal{L}^{(n)}$ is equal to the earlier (multiclass logistic regression) case, and why $\nabla_{\bw_j} \mathcal{L}^{(n)}$ is almost equal to the earlier case.

Like in multiclass logistic regression, you should use intermediate variables $\mathbf{\delta}_j^q$. In addition, you should use intermediate variables $\mathbf{\delta}_j^h = \frac{\partial \mathcal{L}^{(n)}}{\partial h_j}$.

Given an input image, roughly the following intermediate variables should be computed:

$
\log \bq \rightarrow Z \rightarrow \log \bp \rightarrow \mathbf{\delta}^q \rightarrow \mathbf{\delta}^h
$

where $\mathbf{\delta}_j^h = \frac{\partial \mathcal{L}^{(n)}}{\partial \bh_j}$.

Give the equations for computing $\mathbf{\delta}^h$, and for computing the derivatives of $\mathcal{L}^{(n)}$ w.r.t. $\bW$, $\bb$, $\bV$ and $\ba$. 

You can use the convenient fact that $\frac{\partial}{\partial x} \sigma(x) = \sigma(x) (1 - \sigma(x))$.

Using the same definition for $\delta_j^q$ in part 1, we can derive $\delta_j^h$:

$\delta_j^h = \frac{\partial \mathcal{L}^{(n)}}{\partial h_j} = \frac{\partial \mathcal{L}^{(n)}}{\partial \log q_j} \frac{\partial \log q_j}{\partial h_j} = w_{jj}\delta_j^q$

$\frac{\partial \mathcal{L}^{(n)}}{\partial w_j} = \frac{\partial \mathcal{L}^{(n)}}{\partial \log q_j} \frac{\partial \log q_j}{\partial w_j} = \delta_j^q \bh$

$\frac{\partial \mathcal{L}^{(n)}}{\partial \bW} = \bh^T \delta^q$

$\frac{\partial \mathcal{L}^{(n)}}{\partial b_j} = \frac{\partial \mathcal{L}^{(n)}}{\partial \log q_j} \frac{\partial \log q_j}{\partial b_j} = \delta_j^q$

$\frac{\partial \mathcal{L}^{(n)}}{\partial b} = \delta^q$

$\frac{\partial \mathcal{L}^{(n)}}{\partial v_j} = \delta_j^h\frac{\partial h_j}{\partial v_j} = \delta_j^q w_{jj}h_j(1-h_j)x$

$\frac{\partial \mathcal{L}^{(n)}}{\partial v} = w_j \bh(1- \bh)x\delta^q  = \delta_h \bh(1-\bh)x$

$\frac{\partial \mathcal{L}^{(n)}}{\partial a_j} = \delta_j^h \frac{\partial h_j}{\partial a_j} = \delta_j^q w_{jj} h_j(1-h_j)$

$\frac{\partial \mathcal{L}^{(n)}}{\partial a} =  w_{j} \bh(1-\bh)\delta^q = \delta_h \bh(1-\bh)$

As we can see, $\nabla_{\bb} \mathcal{L}^{(n)}$ is equal to the earlier expression, since multi-class logistic regression is similar to the last layer of MLP, except MLP using weighted hidden layer plus bias. That's also why $\nabla_{\bw_j} \mathcal{L}^{(n)}$ is almost similar.

## 2.3. Implement and train a MLP (15 points)

Implement an MLP model with a single hidden layer of **20 neurons**. 
Train the model for **10 epochs**.
Test your implementation for learning rates of 1e-2, 1e-3 and 1e-4 and plot (in one graph) the conditional log-probability of the trainingset and validation set. 

For the best model plot the weights of the first layer for in epoch 0,4 and 9.

- +5 points: Fast, numerically stable, vectorized implementation

```python
# Write all helper functions here
# YOUR CODE HERE
def sigmoid(x):
    return 1/(1 + np.exp(-x))

def compute_grad(x, t, w, b, v, a):
    h = sigmoid(x@v + a)
    logq = h@w + b
    val = np.max(logq)
    logZ = val + np.log(np.sum(np.exp(logq-val)))
    logp = logq - logZ
    ind = np.zeros(logq.shape[1]) 
    ind[t[0]] = 1 # ind has shape(1,10)
    dL_db = ind - np.exp(logq)/np.exp(logZ)
    dL_dw = h.T @ dL_db
    dL_dh = dL_db @ w.T
    dL_dv = x.T @ dL_dh * h * (1-h)
    dL_da = h * (1-h) * dL_dh
    return logp[:,t].squeeze(), dL_dw, dL_db.squeeze(), dL_dv, dL_da.squeeze()
```

```python
# Hidden tests for efficiency
```

```python

def sgd_iter2(x_train, t_train, W, b, V, a, rate):
    # YOUR CODE HERE
    # raise NotImplementedError()
    logp_train = 0
    index = np.arange(0, x_train.shape[0])
    np.random.shuffle(index)
    for i in range(len(index)):
        x = x_train[index[i]:index[i]+1:,]
        t = t_train[index[i]:index[i]+1]
        logp,dw,db,dv,da = compute_grad(x,t,W,b,V,a)
        logp_train += logp
        W += rate * dw
        b += rate * db
        V += rate * dv
        a += rate * da
    return logp_train, W, b, V, a

# Return the loss given parameters w,v,b,a. 
def compute_single_loss(x, t, w, b, v, a):
    h = sigmoid(x@v + a)
    logq = h@w + b
    val = np.max(logq)
    logZ = val + np.log(np.sum(np.exp(logq-val)))
    logp = logq - logZ
    return logp[:,t].squeeze()
  
# Used for compute validation loss    
def compute_whole_loss(X, T, w, b, v, a):
    losses = [compute_single_loss(X[i:i+1,:], T[i], w, b, v, a) for i in range(len(X))]
    return np.sum(losses)

def plot_weight1(weight, num_cols, shape=(28,28)):
    num_digits = weight.shape[1]
    num_rows = int(num_digits/num_cols)
    for i in range(num_digits):
        plt.subplot(num_rows, num_cols, i+1)
        plt.imshow(weight[:,i].reshape(shape), interpolation='none', cmap='Greys')
        plt.colorbar()
        plt.axis('off')
    plt.tight_layout()
    plt.show()
    
    
def plot_loss(train, valid, rate, itter):
    plt.plot(itter, train, color='green', label="Train loss")
    plt.plot(itter, valid, color='blue', label="Valid loss")
    plt.xlabel('Iterations')
    plt.ylabel('Loss')
    plt.title('Learning rate of : '+ str(rate))
    plt.legend()
    plt.show()


def test_mlp(x_train, t_train, x_valid, t_valid, w, b, v, a, rate, iters):
    # YOUR CODE HERE
    v_ys = [] # log-probability of validation set
    ys = [] # log-probability of training set
    xs = [] # x coordinate of plot [1:<=15]
    vs = [] # first_layer weights of num_iters
    i = 1
    while i <= iters:
        logp_train,w,b,v,a = sgd_iter2(x_train, t_train, w, b, v, a, rate)
        valid_logp = compute_whole_loss(x_valid, t_valid, w, b, v, a)
        vs.append(np.copy(v))
        v_ys.append(valid_logp/len(x_valid))
        ys.append(logp_train/len(x_train))
        xs.append(i)
        i+=1
    return w,b,vs,a,ys,v_ys,xs
```

```python
# Write training code here:
# Plot the conditional loglikelihoods for the train and validation dataset after every iteration.
# Plot the weights of the first layer.

# 
# YOUR CODE HERE
np.random.seed(123)
rates = [1e-2, 1e-3, 1e-4]
train_losses = dict()
valid_losses = dict()
plot_v = dict()
for rate in rates:
    # intialize w,b,v,a
    v = np.random.normal(size=(28*28,20), scale=0.001)
    a = np.zeros((20,))
    w = np.random.normal(size=(20,10),scale=0.001)
    b = np.zeros((10,))
    # train
    w,b,v,a,ys,v_ys,xs = test_mlp(x_train, t_train, x_valid, t_valid, w, b, v, a, rate, 10)
    train_losses[rate] = ys
    valid_losses[rate] = v_ys
    plot_v[rate] = v
    plot_loss(ys, v_ys, rate, xs)    
```

Derivative of tanh w.r.t argument:

Name two properties that you would like your activation function to have (one sentence each). Why are they important?


1. Activation function should be continuous and differentiable everywhere, otherwise we can't do backprop.
2. Non-linearity. Otherwise no matter how deep the model goes, it behaviors just like linear regression/classification.


```python
# plot the function and the derivative for the activations sigmoid, tanh and ReLU.

# YOUR CODE HERE

# sigmoid function is defined above
def d_sigmoid(x):
    return sigmoid(x)*(1-sigmoid(x))

def tanh(x):
    return (np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)) 
    
def d_tanh(x):
    return 1 - tanh(x)**2

def relu(x):
    return np.maximum(0,x)

def d_relu(x):
    return np.where(x <= 0, 0, 1)
    
x = np.linspace(-5,5,200)

sigmoid_x = sigmoid(x)
dsigmoid_x = d_sigmoid(x)
plt.plot(x,sigmoid_x, color = 'red', label = "sigmoid")
plt.plot(x,dsigmoid_x, color = 'green', label = "derivative of sigmoid")
plt.title("Sigmoid function")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()

tanh_x = tanh(x)
dtanh_x = d_tanh(x)
plt.plot(x,tanh_x, color = 'red', label = "tanh")
plt.plot(x,dtanh_x, color = 'green', label = "derivative of tanh")
plt.title("tanh function")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()

relu_x = relu(x)
drelu_x = d_relu(x)
plt.plot(x,relu_x, color = 'red', label = "relu")
plt.plot(x,drelu_x, color = 'green', label = "derivative of relu")
plt.title("relu function")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.show()
```
